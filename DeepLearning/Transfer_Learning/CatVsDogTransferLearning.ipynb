```python
train_dir = "../CatsVsDogs/train"
test_dir = "../CatsVsDogs/test"

img_size = 150
```

```python
train_image_list = os.listdir(train_dir)[0:20000]
test_image_list = os.listdir(test_dir)

len(train_image_list), len(test_image_list)
```

```python
train_image_list[15000:]
```

```python
def one_hot_encode(img):
    label = img.split('.')[0]
    if label == 'cat':
        return [1, 0]
    elif label == 'dog':
        return [0, 1]
    
def to_data(data_collection, data_folder, train=True):
    data_df = []
    for img in tqdm(data_collection):
        path = os.path.join(data_folder, img)
        if train:
            label = one_hot_encode(img)
        else:
            label = img.split('.')[0]
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (img_size, img_size))
        data_df.append([np.array(img), np.array(label)])
    shuffle(data_df)
    return data_df

train = to_data(train_image_list, train_dir)
test = to_data(test_image_list, test_dir, False)
```
